```python
# Read .xlsx file and write to .csv file

import pandas as pd

input_file = "/Users/philipacquaye-mensah/Documents/Breast Audit/breast audit data .xlsx"
output_file = "/Users/philipacquaye-mensah/Documents/Breast Audit/breast_audit_data.csv"

# Read the .xlsx file
df = pd.read_excel(input_file, skiprows=5)  # if you need to skip the first two rows

# Write the data to a .csv file
df.to_csv(output_file, index=False)
```

```python
# Print the first 5 rows of the data
print(df.head())
```

```python
# convert to dataframe
df = pd.DataFrame(df)

# print the shape of the dataframe
print(df.shape)

# print the columns of the dataframe
print(df.columns)

# print the data types of the columns
print(df.dtypes)

```

```python
df.head(30)
```

```python
# remove Number column
df = df.drop(['Number '], axis=1)
```

```python
print(df.columns)
```

```python
# rename columns
# remove spaces from column names
df.columns = df.columns.str.replace(' ', '')

# rename last 6 columns
df = df.rename(columns={
    'TypeofUSS(LorRBreastorBoth)': 'UssType',
    'Isthereabreastpicturethroughoutthescan': 'Breast_Icon',
    'Hasafocalareabeenmarked': 'Focal_Marking',
    'FurtherAnnotations': 'Further_Annotations',
    'Furtherannotationdefinedzoneinrelationtonipple?': 'Defined_Zone',
    'Ifstructuresfound,measurementstaken?Y,N&N/A': 'Measurements_Taken'
})
```

```python
df.head(30)
```

```python
# replace Nan values in Measurements_Taken column with NA
df['Measurements_Taken'] = df['Measurements_Taken'].fillna('NA')

df.head(30)
```

```python
df['UssType'].unique()
```

```python
# change categorical values to numerical values
df['UssType'] = df['UssType'].map({'L': 0, 'R': 1, 'B': 2})
df['Breast_Icon'] = df['Breast_Icon'].map({'Y': 1})
df['Focal_Marking'] = df['Focal_Marking'].map({'Y': 1, 'N': 0})
df['Further_Annotations'] = df['Further_Annotations'].map({'Y': 1, 'N': 0})
df['Defined_Zone'] = df['Defined_Zone'].map({'Y': 1, 'N': 0})
df['Measurements_Taken'] = df['Measurements_Taken'].map({'Y': 1, 'N': 0, 'NA': 2})

df.head(30)
```

```python
# find the unique values in the UssType column
df['UssType'].unique()
```

```python
# check how many nan values in UssType column
df['UssType'].isnull().sum()
```

```python
# show the rows with nan values in UssType column
df[df['UssType'].isnull()]
```

```python
# replace nan values in UssType column row by row
df.loc[8, 'UssType'] = 0
df.loc[39, 'UssType'] = 0
df.loc[66, 'UssType'] = 0
df.loc[80, 'UssType'] = 0
df.loc[82, 'UssType'] = 2
df.loc[91, 'UssType'] = 1

# replace nan values in Breast_Icon column row by row
df.loc[39, 'Breast_Icon'] = 1.0
df.loc[82, 'Breast_Icon'] = 1.0

# replace nan values in Measurements_Taken column row by row
df.loc[66, 'Measurements_Taken'] = 2.0
```

```python
# calculate the percentage of 0.0 in UssType column
percentage = (df['UssType'].value_counts()[0.0] / df.shape[0]) * 100
print(f'{percentage:.2f}%')
```

```python
# calculate the percentage of 1.0 in UssType column
percentage = (df['UssType'].value_counts()[1.0] / df.shape[0]) * 100
print(f'{percentage:.2f}%')
```

```python
# calculate the percentage of 2.0 in UssType column
percentage = (df['UssType'].value_counts()[2.0] / df.shape[0]) * 100
print(f'{percentage:.2f}%')
```

```python
# visualise the distribution of the UssType column using a bar chart with the percentage of each category
# Lable 0.0 as Left, 1.0 as Right and 2.0 as Both
import matplotlib.pyplot as plt
df['UssType'] = df['UssType'].map({0: 'Left', 1: 'Right', 2: 'Both'})

# add the percentage of each category to the plot at the top of each bar
ax = df['UssType'].value_counts().plot(kind='bar')
for i in ax.patches:
    ax.text(i.get_x() + i.get_width() / 2, i.get_height() + 0.5,
            f'{(i.get_height() / df.shape[0] * 100):.2f}%', ha='center')
plt.xlabel('UssType')
plt.ylabel('Frequency')
plt.title('Type of Breast Ultrasound Scan')
plt.show()
```

```python
# check Breast_Icon column unique values
df['Breast_Icon'].unique()

# check how many nan values in Breast_Icon column
df['Breast_Icon'].isnull().sum()

# replace nan values in Breast_Icon column with 1.0
df['Breast_Icon'] = df['Breast_Icon'].fillna(1.0)

# calculate the percentage of 1.0 in Breast_Icon column
percentage = (df['Breast_Icon'].value_counts()[1.0] / df.shape[0]) * 100
print(f'{percentage:.2f}%')
```

```python
# calculate the percentage of 1.0 in Focal_Marking column
percentage = (df['Focal_Marking'].value_counts()[1.0] / df.shape[0]) * 100
print(f'{percentage:.2f}%')

# check how many nan values in Focal_Marking column
df['Focal_Marking'].isnull().sum()
```

```python
# check how many nan values in Further_Annotations column
df['Further_Annotations'].isnull().sum()

# show the rows with nan values in Further_Annotations column
df[df['Further_Annotations'].isnull()]

# replace nan values in Further_Annotations column row by row
df.loc[40, 'Further_Annotations'] = 0.0

# calculate the percentage of 0.0 in Further_Annotations column
percentage = (df['Further_Annotations'].value_counts()[0.0] / df.shape[0]) * 100
print(f'{percentage:.2f}%')

# calculate the percentage of 1.0 in Further_Annotations column
percentage = (df['Further_Annotations'].value_counts()[1.0] / df.shape[0]) * 100
print(f'{percentage:.2f}%')
```

```python
# calculate the percentage of 1.0 in Further_Annotations column with 1.0 in Defined_Zone column
percentage = df[(df['Further_Annotations'] == 1.0) & (df['Defined_Zone'] == 1.0)].shape[0] / df.shape[0] * 100
print(f'{percentage:.2f}%')
```

```python
# visualise all results in a pie char
# create a list of the values to be visualised
values = [df['Further_Annotations'].value_counts()[0.0], df['Further_Annotations'].value_counts()[1.0],
          df[(df['Further_Annotations'] == 1.0) & (df['Defined_Zone'] == 1.0)].shape[0]]
labels = ['No Further Annotations', 'Further Annotations', 'Further Annotations with Defined Zone']

# create a pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Further Annotations')
plt.show()
```
